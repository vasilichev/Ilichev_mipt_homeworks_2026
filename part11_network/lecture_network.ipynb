# Лекция: Веб-технологии

В 70-х годах прошлого века помимо мейнфреймов (больших компьютеров размером с комнату) стали появляться относительно небольшие компьютеры. Устройств становилось всё больше, потребность в новых компьютерных и сетевых архитектурах стала ещё острее.

Ключевые вехи:
- В 1974 году представили концепцию протокола TCP (Transmission Control Protocol)
- В 1978 году произошло разделение: TCP стал отвечать за надежную доставку данных, а новый протокол IP (Internet Protocol) взял на себя маршрутизацию пакетов данных.
- 1984 г. - Создание системы доменных имён для удобной адресации (DNS - Domain Name System)
- 1989-1991 гг. - Создание World Wide Web (Тим Бернерс-Ли)

Когда мы хотим отправить какие-то данные, байты с нашей информацией на каждом слое упаковываются в специальные фреймы с дополнительными заголовками, проходят через сеть и потом так же последовательно разбираются. Главное здесь, что появился полный набор протоколов и их реализаций для безопасной и надёжной передачи данных в сети с большим количеством устройств.

DATA здесь представляется байтами, но вот вопрос - а что и когда отправлять?

Допустим, у нас есть нужный IP-адрес и некоторое сообщение. Как принимающая сторона поймёт, что ей отправили и как это читать? Вы отправили текст или строку? Ожидаете ли вы ответа?

_Примечание: у Андрея Созыкина есть классные [курсы](https://www.youtube.com/playlist?list=PLtPJ9lKvJ4ojPWFLuUz6g8c73Ta45bUN8) по компьютерным сетям. Очень рекомендую к ознакомлению._

Представьте, что вам выдали рации для связи друг с другом.

По идее, совсем необязательно знать, как именно они доставляют сигналы, но очень важно выработать правила выхода в эфир:

- Кондор вызывает Суслика, **приём**.
- Кондор, Кондор, Суслик на связи, **приём**.
- Суслик, посылка отправлена, **конец связи**.

Слово "приём" здесь обозначает конец сообщения и ожидание ответа, "конец связи" - окончание коммуникации.

К концу 80-х нужные способы доставки сообщений уже были, именно там и началось формирование "правил".

# WWW и HTML

<img src="./_resources/www.jpg" height=500>

В 1989 году Тим Бернерс-Ли придумал **World-Wide-Web** - связывать страницы гиперссылками в одну большую паутину, к которой у всех будет доступ. Страничка до сих пор [доступна](https://info.cern.ch/hypertext/WWW/TheProject.html))

Страницы были текстовыми, но уже со специфической разметкой на языке **HTML** (HyperText Markup Language). Суть проста - принимающая сторона (браузер) сама отвечает за рендеринг страницы (правильное отображение гиперссылок, таблиц, стилей теста).

Чтобы адреса к ресурсам имели единообразный формат, в 1990 году создатели Всемирной паутины разработали специальный стандарт, определяющий, как должны выглядеть адреса ресурсов. Этот стандарт называется **URL** (Uniform Resource Locator)

А уже ближе к 1991 году на свет появилась реализация HTTP 0.9...

# HTTP и HTTPS

<img src="./_resources/http_meme.jpg" height=600>

**Кто этот ваш HTTP?**

**НТТР** (англ. HyperText Transfer Protocol - "протокол передачи гипертекста») - протокол прикладного уровня передачи данных, изначально в виде гипертекстовых документов в формате HTML. В настоящий момент используется для передачи произвольных данных.

## Как работает HTTP?

Давайте для примера возьмём один URL: https://server.org/test/test.html?param1=a&param2=b

**https** - протокол

**server.org** - доменное имя сервера 

**/test/test.html** - путь к ресурсу

**param1=a&param2=b** - параметры запроса

Примечание: если написать **param1=a&param1=b** - будет считано только последнее значение.

<img src="./_resources/domain_names.png">

Наш HTTP-запрос будет работать следующим образом:

1. Браузер анализирует URL и извлекает доменное имя (server.org.)
2. Браузер обращается к DNS серверу и получает от него IP-адрес, к которому нужно обратиться (например, 77.88.44.242)
3. Между клиентом и сервером создаётся TCP-соединение (Стучимся в 77.88.44.242:443 - для HTTPS закреплён порт 443, для HTTP - 80)
4. Браузер формирует HTTP-запрос
5. Через TCP-соединение запрос отправляется на сервер
6. Дожидаемся HTTP-ответа от сервера
7. Закрываем TCP-соединение

8. Обрабатываем ответ. PROFIT.

## Типы HTTP-запросов

Запросы могут быть не только на чтение. HTTP [имеет](https://developer.mozilla.org/ru/docs/Web/HTTP/Reference/Methods) специальные методы, которые позволяют серверу понять, что именно вы хотите сделать:
- GET - получение ресурса
- POST - отправка данных на сервер (обычно создание ресурса)
- PUT - отправка данных (обычно редактирование, редактируются все данные ресурса)
- PATCH - отправка данных (тоже редактирование, но редактируется часть ресурса)
- DELETE - удаление ресурса

А также есть методы:
- HEAD - апрашивает ресурс так же, как и метод GET, но без тела ответа
- OPTIONS - используется для описания параметров соединения с ресурсом
- TRACE - выполняет вызов возвращаемого тестового сообщения с ресурса
- CONNECT - устанавливает "туннель" к серверу, определённому по ресурсу

## Типы HTTP-ответов

Представьте, мы отправили запрос, а нам вернулась строка. Это текст документа? Это текст ошибки? Как понять, успешен ли запрос?

Для этого HTTP-ответ содержит специальный код ответа. Есть 5 групп ответов:
- 1хх - информационные
- 2хх - успех
- 3хх - перенаправления и редиректы
- 4хх - ошибка на стороне клиента
- 5хх - ошибка на стороне сервера

Полный список есть в [документации](https://developer.mozilla.org/en-US/docs/Web/HTTP/Reference/Status).

Чаще всего возникают коды:
* 200 - операция прошла успешно
* 201 - запись успешно создана
* 400 - некорректный запрос
* 401 - пользователь не прошёл обязательную аутентификацию
* 403 - нет прав на обращение к ресурсу
* 404 - ресурс не найден
* 422 - некорректные данные в запросе
* 429 - слишком много обращений к серверу от клиента
* 500 - внутренняя ошибка сервера
* 502 - сервер получил ошибку от другого сервера
* 503 - сервер не доступен (например, перегружен)
* 504 - превышено время ожидания ответа

## HTTP 0.9

- Единственный метод — GET
- Отсутствие заголовков
- Передача только HTML-документов
- Соединение закрывалось сразу после отправки ответа
- Отсутствие кодов состояния — сервер мог только вернуть документ или сообщение об ошибке

Типичный запрос в HTTP/0.9 выглядел так:

GET /index.html

А ответ представлял собой непосредственно HTML-документ:

```html
<html>
<body>
<h1>Hello, World!</h1>
</body>
</html>
```

## HTTP/1.0

- Добавление HTTP-заголовков для метаданных
- Поддержка различных типов контента через Content-Type
- Новые методы — POST и HEAD
- Введение кодов состояния (200 OK, 404 Not Found и др.)
- Поддержка кэширования через заголовки If-Modified-Since и Last-Modified
- Базовая аутентификация

## HTTP/1.1

В 1997 году появился HTTP/1.1. Основные улучшения:

- Постоянные соединения (Connection: keep-alive) — возможность использовать одно TCP-соединение для нескольких запросов
- Конвейерная обработка запросов — возможность отправить несколько запросов без ожидания ответов
- Новые методы — PUT, DELETE, TRACE, OPTIONS
- Улучшенное кэширование с новыми директивами и заголовками
- Частичные запросы (Range) для возобновления загрузки
- Виртуальный хостинг — поддержка нескольких доменов на одном IP-адресе через заголовок Host
- Согласование содержимого — выбор предпочтительного представления ресурса
- Сжатие с помощью Content-Encoding

## HTTP/2

HTTP/2 - стандартизирован в мае 2015 года, вырос из протокола SPDY от Google.

Ключевые инновации HTTP/2 включают:

- Мультиплексирование — возможность отправлять множество запросов и получать ответы асинхронно в рамках одного соединения
- Бинарный формат — замена текстового формата протокола на более компактный и эффективный бинарный
- Сжатие заголовков через алгоритм HPACK, уменьшающий избыточность
- Серверный push — возможность для сервера отправлять ресурсы клиенту до их запроса
- Приоритизация потоков — указание важности запросов для оптимального распределения ресурсов

- Одно соединение на домен вместо множества параллельных

## HTTP/3

HTTP/3 использует новый транспортный протокол QUIC вместо TCP. QUIC был разработан Google и впервые внедрен в Chrome в 2013 году, а затем стандартизирован IETF. HTTP/3 официально стандартизирован в RFC 9114 в июне 2022 года.

Фундаментальные преимущества HTTP/3 и QUIC:

- Устранение блокировки начала очереди на транспортном уровне — потеря одного пакета затрагивает только один поток данных
- Уменьшенное время установки соединения — в идеальном случае всего один раунд (0-RTT)
- Улучшенная работа в нестабильных сетях — особенно в мобильных сценариях с частой сменой сетей
- Встроенное шифрование — TLS интегрирован в протокол
- Работа на уровне UDP — обход проблем с промежуточными устройствами, модифицирующими TCP

- Миграция соединения — возможность сохранения сессии при смене IP-адреса

# В чём отличие HTTP от HTTPS?

На заре интернета информация была в основном публичной, поэтому вопрос безопасности первое время стоял не так остро.

Однако перед атакой Man-in-the-Middle обычный HTTP бессилен. Так появился HTTPS.

HTTPS (HyperText Transfer Protocol Secure) — это надстройка: тот же самый HTTP, но запущенный поверх криптографического протокола SSL/TLS (Secure Sockets Layer / Transport Layer Security). Сейчас используется преимущественно TLS, но название SSL исторически закрепилось. Представьте себе конверт: HTTP — это письмо, а TLS — это запечатанный и подписанный конверт, который защищает содержимое и подтверждает отправителя.

Ключевое отличие: если HTTP работает напрямую по TCP на порту 80, то HTTPS сначала устанавливает зашифрованный канал (обычно порт 443), и только внутри этого канала начинает работать привычный HTTP.


При установке HTTPS-соединения происходит процедура, называемая «рукопожатием» (TLS handshake). За пару раундов обмена сообщениями клиент и сервер:

1. Договариваются об алгоритме шифрования (cipher suite).

2. Сервер предоставляет цифровой сертификат (как удостоверение личности), подписанный доверенным центром (Certificate Authority, CA).

3. С помощью асимметричной криптографии (открытый/закрытый ключ) стороны безопасно согласовывают временный симметричный ключ для сессии.

После этого весь HTTP-трафик (заголовки, куки, тело запроса/ответа) шифруется симметричным алгоритмом (например, AES). Это быстро для процессора и надёжно.

**Что даёт HTTPS (три кита безопасности)**

В отличие от чистого HTTP, протокол HTTPS гарантирует три важных свойства:

- Шифрование (Confidentiality): перехваченные пакеты превращаются в белый шум. Даже зная весь трафик, злоумышленник не прочитает пароль.

- Целостность (Integrity): любое случайное или намеренное изменение данных в пути будет обнаружено — соединение разорвётся.

- Аутентификация (Authentication): браузер может проверить, что он действительно общается с нужным сервером (например, google.com), а не с подделкой. Зелёный замочек в адресной строке — визуальный признак этого.

# REST - Архитектурный стиль

REST (Representational State Transfer) — это набор принципов и соглашений о том, как создавать и работать с веб-запросами.

По сути, REST предлагает простой и понятный способ взаимодействия между клиентом и сервером, следуя которому разработчики облегчают себе и другим работу с веб-сервисами.

Основные принципы REST:

1. **Клиент-серверная архитектура**

    Что это значит? Система разделена на две отдельные части:

    Клиент — отправляет запросы к серверу.

    Сервер — принимает эти запросы и обрабатывает их.

    Зачем это нужно? Разделение позволяет клиенту и серверу развиваться независимо друг от друга, упрощает масштабирование и поддержку системы.
    Пример:
    - Веб-приложение: Браузер (клиент) запрашивает данные у веб-сервера для отображения веб-страницы.
    - Пример неклиент-серверной архитектуры: Блокчейн-сеть, где каждый узел одновременно является и клиентом, и сервером, и все узлы равноправны.


2. **Отсутствие состояния (Stateless)**:

    Что это значит? Каждый запрос от клиента к серверу должен содержать всю необходимую информацию для его обработки. Сервер не сохраняет информацию о состоянии клиента между запросами.

    Зачем это нужно? Упрощает серверную архитектуру, облегчает масштабирование и делает систему более устойчивой.

    Пример:
    - При каждом запросе клиент отправляет токен аутентификации, поскольку сервер не "помнит" предыдущие запросы.
    - Противоположность: В интернет-магазине корзина покупок сохраняется на сервере между запросами, что нарушает принцип Stateless.


3. **Кэширование**:

    Что это значит? Ответы сервера могут быть помечены как кэшируемые, позволяя клиенту или промежуточным узлам сохранять их и повторно использовать при последующих запросах.

    Зачем это нужно? Снижает нагрузку на сервер, уменьшает задержки и повышает производительность системы.

    Пример:
    - Статические ресурсы: Изображения, стили и скрипты веб-сайта кэшируются в браузере пользователя.
    - API-ответы: Часто запрашиваемые данные, такие как курсы валют, могут кэшироваться на стороне клиента.


4. **Единообразие интерфейса**:

    Что это значит? Использование стандартных методов и форматов для взаимодействия между клиентом и сервером.

    Зачем это нужно? Упрощает разработку, позволяет различным системам легко взаимодействовать друг с другом.
    
    Пример:
    - HTTP-методы: Использование методов GET (для получения данных), POST (для создания), PUT (для обновления) и DELETE (для удаления).
    - Самоописываемые сообщения: Сервер возвращает не только данные, но и информацию о том, какие действия можно выполнить дальше.
    - Гипермедиа как движок состояния приложения (HATEOAS): Клиент может навигировать по API, используя ссылки, предоставленные в ответах сервера.


5. **Многоуровневая система (Layered System)**:

    Что это значит? Архитектура может состоять из нескольких уровней (прокси-серверы, шлюзы, балансировщики нагрузки), и каждый компонент взаимодействует только со своим непосредственным соседом.

    Зачем это нужно? Повышает масштабируемость, безопасность и управляемость системы.

    Пример:
    - Клиент не знает, общается ли он напрямую с сервером или через промежуточные узлы.
    - Промежуточные кэши: Могут кэшировать ответы для улучшения производительности.


6. **Код по требованию (Code on Demand, опционально)**:

    Что это значит? Сервер может передавать клиенту исполняемый код, который расширяет его функциональность.

    Зачем это нужно? Позволяет обновлять и расширять возможности клиента без его модификации.

    Пример:
    - Веб-приложения: Браузер загружает и выполняет JavaScript-код, предоставленный сервером, для динамического взаимодействия с пользователем.

    - Мобильные приложения: Получают скрипты или конфигурации с сервера для изменения поведения без обновления приложения.


# От «голого» сокета до асинхронного конвейера

Представьте себе веб-сервер начала 2000-х. Это такая программа (Apache, Nginx), которая умеет слушать порт 80, парсить HTTP-заголовки и отдавать статические файлы с диска. Но как только запрос требует «подумать» — заглянуть в базу, сформировать PDF, отправить email — сервер бессилен. Он не знает ваш Python-код.
Первый прорыв: CGI и появление «мостиков»

Родилась идея общего шлюзового интерфейса (CGI). Сервер при получении запроса порождает новый процесс (например, python script.py), передаёт туда переменные окружения и тело запроса, ждёт вывода результата. Работает, но ужасно неэффективно: создание процесса на каждый клик — это дорого.

На смену пришли встроенные модули (mod_python) и, наконец, WSGI (Web Server Gateway Interface) для Python. Это простой синхронный контракт: веб-сервер (или сервер приложений вроде Gunicorn) вызывает вашу функцию, передавая ей словарь environ и callback start_response, а ваша функция возвращает итератор с байтами ответа.

```python
# Минимальное WSGI-приложение
def app(environ, start_response):
    start_response('200 OK', [('Content-Type', 'text/plain')])
    return [b'Hello, world!']
```

WSGI решил главную проблему: сделал веб-фреймворки (Django, Pyramid, Flask) независимыми от сервера. Можно писать код, не думая, запустишь ты его через Gunicorn, uWSGI или Apache mod_wsgi.
Но WSGI столкнулся с потолком синхронности.

В современном вебе слишком много ситуаций, где процессор не работает, а соединение висит вхолостую:

- Long polling (ждём события на сервере)

- WebSockets (двусторонний обмен поверх HTTP)

- Server-Sent Events (SSE)

- Долгие запросы к БД или внешним API

В синхронной WSGI-модели серверный поток (или процесс) блокируется до тех пор, пока ваша функция не вернёт ответ. Пока один клиент ждёт ответ от медленной базы, этот поток не может обслужить сотню других. Приходится запускать тысячи потоков/процессов, что прожорливо по памяти и плохо масштабируется.

Тогда зачем понадобились фреймворки, такие как Flask, Django или FastAPI?

Каждый из них — ответ на вопрос: «Как мне удобно писать веб-приложения, не думая о низкоуровневых сокетах, но с учётом ограничений синхронной/асинхронной модели?»

Позже был принят стандарт ASGI (Asynchronous Server Gateway Interface). ASGI — это WSGI, но с поддержкой асинхронности и нескольких протоколов.

WSGI умел только HTTP и синхронный запрос-ответ. ASGI моделирует приложение как асинхронную функцию с тремя параметрами: scope (информация о соединении, как environ), receive (асинхронное чтение входящих событий), send (асинхронная отправка). Это позволяет:

- Обрабатывать WebSocket (непрерывный обмен сообщениями)

- Держать тысячи долгоживущих соединений с минимальным расходом памяти

- Иметь фоновые задачи в рамках того же цикла событий

Теперь архитектура выглядит так:

```
Клиент (HTTP/WebSocket) → ASGI-сервер (Uvicorn, Daphne) → ASGI-приложение (FastAPI, Quart, Starlette)
```

Старые WSGI-фреймворки (Flask, Django до 3.0) можно запускать под ASGI-сервером, но они будут работать синхронно, оборачивая каждый вызов в поток. ASGI раскрывается на 100% только с асинхронными фреймворками.

# FastAPI

FastAPI — современный веб-фреймворк для Python (асинхронный, на основе Starlette и Pydantic).

Главные черты:
- высокая производительность (на уровне NodeJS/Go)
- автоматическая генерация OpenAPI и документации (Swagger, ReDoc)
- встроенная валидация данных через типы Python
- асинхронные эндпоинты.

Используется для создания REST API и микросервисов.

Вот пример из документации: "[Первые шаги](https://fastapi.tiangolo.com/tutorial/first-steps/)"

Давайте напишем небольшой сервер, у которого будет только одна ручка (эндпойнт), через которую клиент может узнать, жив ли сервер.

````python
import uvicorn
from fastapi import FastAPI, Response, status

app = FastAPI()


@app.get("/api/v1/healthcheck")
def healthcheck() -> Response:
    """Проверить, что сервер активен."""
    return Response(status_code=status.HTTP_200_OK)


def main() -> None:
    """Запустить сервер."""
    uvicorn.run(app)


if __name__ == "__main__":
    main()
````

Разберём, что здесь происходит.

1. Создали экземпляр нашего FastAPI приложения. Он будет хранить всю информацию о наших ресурсах: `app = FastAPI()`

2. Объявили функцию `def healthcheck` и задекорировали её в `@app.get("/api/v1/healthcheck")`

   Таким образом мы явно зарегистрировали нашу функцию, как обработчик HTTP GET запросов к пути /api/v1/healthcheck.

   Регистрировать эндпойнты на один и тот же путь - плохо. FastAPI возьмёт только первый из них.

3. Запустили ASGI-сервер. (Обычно его запускают отдельно, не в коде приложения)

Теперь откроем `http://localhost:8000/docs`. 

Открывшийся интерфейс называется `Swagger`. Это инструмент, который позволяет на основе некоторой спецификации (формальное описание API) выполнять запросы к некоторому серверу от имени клиента.

`Swagger` строится на основе спецификации от `OpenAPI`. Давайте взглянем на спецификацию на примере нашего сервера. Перейдем в `Swagger`. Слева вверху можно увидеть гиперссылку на `/openapi.json` - перейдем по ней. Увидим следующее:

# Пример побольше

Давайте сделаем минимальный пример интернет-магазина. Пусть он имеет сущность **Товар** и поддерживает отображение **Отзывов** о товаре.

````python
from dataclasses import dataclass


@dataclass(frozen=True, slots=True)
class Good:
    id: int
    title: str
    description: str


@dataclass(frozen=True, slots=True)
class Review:
    id: int
    good_id: int
    body: str


goods: dict[int, Good] = {
    1: Good(1, "Шампунь Жумайсынба", "Не можете здороваться с братками? Наш шампунь вам поможет!"),
    2: Good(2, "Слон", "А ты купи слона"),
    3: Good(3, "Бананы", "Шучу, не бананы"),
}

reviews: dict[int, Review] = {
    1: Review(1, 1, "От этого шампуня волосы дыбом! Куплю ешё!"),
    2: Review(2, 1, "Доставили быстро, ещё не открывал"),
    3: Review(3, 2, "Бананы хорошие, но кот напрудил мне в тапки, поэтому 4"),
}
````

````python
from fastapi import FastAPI, HTTPException

app = FastAPI(
    title="API магазина ДомашниеЯгоды",
    version="0.1.0",
)

@app.get("/goods/{good_id}")  # Это PATH-параметр.
def get_good(good_id: int):
    ...
    # /good

@app.get("/goods/{good_id}/search")
def get_comments_for_good(good_id: int, phrase: str): # good_id - Path, phrase - Query-параметр
    ...
````

Про параметры можно почитать [здесь](https://fastapi.tiangolo.com/tutorial/path-params/) и [здесь](https://fastapi.tiangolo.com/tutorial/query-params/).

А как отдавать что-то посложнее примитивных типов? Для этого FastAPI сразу тянет с собой библиотеку Pydantic.

Она позволяет легко описывать схемы ввода/вывода для наших приложений.

````python
from pydantic import BaseModel, Field

# В самом базовом случае можно просто оставить атрибуты и их типы

class GoodSchema(BaseModel):
    id: int
    title: str
    description: str

# Но pydantic ещё представляет класс Field, в котором можно описать ограничения атрибута или доп. информацию

class ReviewSchema(BaseModel):
    id: int = Field(description="Идентификатор отзыва")
    good_id: int = Field(description="Идентификатор товара")
    body: str = Field(description="Текст отзыва")

# А вот так можно использовать вложенные схемы

class ListReviewSchema(BaseModel):
    """Список отзывов на товар"""  # Кстати, этот комментарий будет виден в OpenAPI-спецификации.
    reviews: list[ReviewSchema]
````

````python
@app.get("/goods/{good_id}")
def get_good(good_id: int) -> GoodSchema:
    """
    Возвращает товар по идентификатору
    """
    good = goods.get(good_id)
    if good is None:
        raise HTTPException(status_code=404, detail="Good not found!")
    return GoodSchema(**asdict(good))


@app.get("/goods/{good_id}/reviews")
def get_reviews_for_good(good_id: int, phrase: str | None = None) -> ListReviewSchema:
    """
    Возвращает все отзывы для товара с опциональным поиском по фразе
    """
    needle = phrase or None
    found_reviews = (r for r in reviews.values() if r.good_id == good_id and (needle is None or needle in r.body))
    return ListReviewSchema(
        reviews=[ReviewSchema(**asdict(review)) for review in found_reviews],
    )


@app.post("/goods/{good_id}/reviews")
def create_review(
    good_id: Annotated[int, Path(title="The ID of good", ge=0)],
    create_review: Annotated[CreateReviewSchema, Body()],
) -> ReviewSchema:
    """
    Создаёт новый отзыв
    """
    if good_id not in goods:
        raise HTTPException(status_code=404, detail="Good not found!")
    new_review = Review(max(reviews.keys(), default=0) + 1, good_id, **create_review.model_dump())
    reviews[new_review.id] = new_review
    return ReviewSchema(**asdict(new_review))
````

Вообще, FastAPI сам понимает, какие параметры лежат в path, а какие в query.

Но связка Annotated[тип, Path], Annotated[тип, Query], Annotated[тип, Body] позволяет добавить валидации, комментарии и примеры. 

И ещё один аспект: когда эндпойнтов становится много и/или они относятся к разным группам задач, их принято группировать в роутеры

# Зависимости в FastAPI

В FastAPI есть свой механизм для Dependency Injection (Инъекция зависимостей) - Depends. См. [документацию](https://fastapi.tiangolo.com/reference/dependencies/)

Допустим, у вас есть генератор, который по запросы выдаёт новую сессию для подключения к базе данных.

Чтобы каждый раз не обращаться напрямую, можно оборачивать его в Depends. Так к началу выполнения основной функции мы уже получим необходимые объекты:

````python
from typing import Annotated

from fastapi import Depends, FastAPI, Header, HTTPException

app = FastAPI()


def verify_token(x_token: Annotated[str, Header()]):
    if x_token != "fake-super-secret-token":
        raise HTTPException(status_code=400, detail="X-Token header invalid")


def verify_key(x_key: Annotated[str, Header()]):
    if x_key != "fake-super-secret-key":
        raise HTTPException(status_code=400, detail="X-Key header invalid")
    return x_key


@app.get("/items/", dependencies=[Depends(verify_token), Depends(verify_key)])
def read_items():
    return [{"item": "Foo"}, {"item": "Bar"}]

def some_method():
    yield 10

@app.put("/items/")
def read_items(some_value = Depends(some_method)):
    return [{"item": "Foo"}, {"item": "Bar"}]
````

# Middleware и CORS

**Middleware** — это слой «обёртки» вокруг приложения: каждый входящий HTTP-запрос и исходящий ответ проходят через цепочку функций или компонентов. Middleware может добавлять заголовки, логировать запросы, проверять аутентификацию, измерять время, перехватывать ошибки, ограничивать частоту запросов и т.д. Важно понимать порядок выполнения: обычно сначала вызывается «внешний» middleware при входе запроса, затем следующий уровень, пока не дойдём до маршрута (handler); при возврате ответа порядок обратный — ответ «всплывает» через те же слои. Таким образом, middleware — это единообразный способ вынести сквозную логику из отдельных эндпоинтов.

````python
from starlette.middleware.cors import CORSMiddleware

app = FastAPI()
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)
````

**CORS** (Cross-Origin Resource Sharing) — механизм браузера, который решает, может ли веб-страница с одного origin (схема + хост + порт) делать запросы к API на другом origin. Без соответствующих заголовков на стороне сервера браузер заблокирует «непростые» запросы или скроет ответ из JS (preflight для методов вроде POST с нестандартными заголовками и т.д.). CORS ослабляет политику одного источника для доверенных фронтендов.

В FastAPI типичный путь — подключить CORSMiddleware из Starlette: указать список разрешённых источников (allow_origins), методы, заголовки, можно ли передавать cookies (allow_credentials). CORS касается только браузеров; запросы из Postman, сервер-сервер или мобильное приложение не подчиняются той же политике.

**Ситуация**

Фронтенд открыт как https://app.example.com, API — https://api.example.com (другой origin). Допустим, в FastAPI нет CORSMiddleware и вы нигде не выставляете заголовки Access-Control-*.

Пользователь нажимает кнопку «Загрузить профиль», браузер делает, например:

```text
GET /users/me HTTP/1.1
Host: api.example.com
Authorization: Bearer …
```

Сервер получает запрос, маршрут отрабатывает, в логах видно 200 и тело JSON — с точки зрения сервера всё ок.

**Что видит браузер**

Политика одного источника не даёт JavaScript’у прочитать ответ с другого origin, если сервер явно не разрешил это через CORS. В консоли разработчика (Chrome и др.) появится ошибка в духе:

«Access to fetch at 'https://api.example.com/users/me' from origin 'https://app.example.com' has been blocked by CORS policy: No 'Access-Control-Allow-Origin' header is present on the requested resource.»
Во вкладке Network запрос может отображаться как failed или с красной пометкой; для «простого» GET иногда видно статус, но в коде fetch/axios приходит ошибка, данных из ответа нет.

То есть проблема не в том, что API «не работает», а в том, что браузер не отдаёт ответ странице без нужных заголовков.
