## Frozenset

Давайте рассмотрим для начала ещё один встроенный  тип данных, о котором мало кто вспоминает. Это то же самое множество, но его нельзя изменять. Как и обычное множество, frozenset может хранить только ХЭШИРУЕМЫЕ типы данных. Самый важный плюс frozenset - его можно использовать вместо ключей словаря, т.к frozenset - неизменяемый (immutable) тип данных. Давайте посмотрим на его использование

```python
fs = frozenset([1,2,3,2,3,4,5,1])
```

```python
fs
```

```python
t = ('a', 'b', 'b', 'b', True, 1, 1.0, 10, (1,2,3), 'a')
fs = frozenset(t)
fs
```

```python
frozenset()
```

```python
fs = frozenset([1,2,3,[1,2,3], {6,5,5}])
```

```python
# Операции для frozenset доступны те же самые, что и для обычного множества, кроме добавления и удаления элементов.

s = {'abc', 10, -5.3, True}
fs = frozenset(('a','b','c', 10, False, True))
```

```python
print(s | fs)
print(fs | s)
print(fs.union(s))
```

```python
print(s & fs)
print(fs & s)
print(fs.intersection(s))
```

```python
print(s - fs) # Остаются элементы s которых нет в fs
print(fs - s) # Элементы fs, которых нет в s.
print(fs.difference(s))
```

```python
print(s ^ fs)
print(fs ^ s)
print(fs.symmetric_difference(s))
```

Также доступны операции сравнения и вообще все операции, которые доступны для множества. Также давайте посмотрим, каким образом можно использовать frozenset в качестве ключа словаря.

```python
d_bad = {s: "set", fs: "frozenset"}
```

```python
d_good = {fs: "frozenset"}
d_good
```

```python
fs = frozenset(('a','b','c', 10, False, True))
fs1 = frozenset(('a','b','c', 10, False, True))
hash(fs) == hash(fs1)
```

## Emun. Перечисления

Очень популярная библиотека, которая дает возможность создавать перечисления в виде класса. Например, мы хотим создать перечисления для обработки разных форматов изображений. 

```python
from enum import Enum


class ImageExtensionSimple:
    WEBP = 'wepb'
    AVIF = 'avif'
    JXL = 'jxl'
    SVG = 'svg'


class ImageExtension(Enum):
    WEBP = 'wepb'
    AVIF = 'avif'
    JXL = 'jxl'
    SVG = 'svg'
```

```python
ImageExtensionSimple.WEBP
```

```python
ImageExtensionSimple.WEBP = 0
```

```python
ImageExtensionSimple.WEBP
```

```python
ImageExtension.WEBP
```

```python
ImageExtension.WEBP = 0
```

Важное замечание для класса перечислений (Enum): мы не можем менять атрибуты этого класса на другие значения. Если мы создали класс перечисления, то атрибут становится константой.

```python
class Color(Enum):
    red = (255, 0, 0)
    green = (0, 255, 0)
    blue = (0, 0, 255)
```

```python
# Если мы хотим использовать только уникальные значения enum

from enum import unique


@unique
class ExtendedColorWithAlias(Enum):
    red = (255, 0, 0)
    green = (0, 255, 0)
    blue = (0, 0, 255)
    gray = 'gray'
    black = 255
    white = 255
```

```python
@unique
class ExtendedColorWithAlias(Enum):
    red = (255, 0, 0)
    green = (0, 255, 0)
    blue = (0, 0, 255)
    gray = 'gray'
    black = 0
    white = 255
```

```python
Color.red == ExtendedColor.red
```

```python
def processing_image(image: bytes, image_ext: ImageExtension = ImageExtension.WEBP):
    match image_ext:
        case ImageExtension.WEBP:
            print(f"Fast processing WEBP {image=}")
        case ImageExtension.AVIF:
            print(f"Processing AVIF {image=}")
        case ImageExtension.JXL:
            print(f"Modern processing JXL {image=}")
        case ImageExtension.SVG:
            print(f"Slow processing SVG graphics {image=}")
        case _:
            raise ValueError(f"We can process only {[p.name for p in ImageExtension]} image extensions")
```

```python
processing_image(b'webp')
processing_image(b'avif', ImageExtension.AVIF)
processing_image(b'jxl', ImageExtension.JXL)
processing_image(b'svg svg svg', ImageExtension.SVG)
processing_image(b'unknown', 'UNKNOWN')
```

Как мы можем видеть, основное назначение использования перечисления это повышение читаемости текста программы. Любой программист, увидевший функцию выше, сможет понять, что она делает.

```python
processing_image(b'webp', 'webp')
```

Но почему теперь мы и видим обработку изображения в WEBP, если webp у нас указано в перечислении. Мы с вами уже говорили, что перечисления используют для сравнения атрибуты классов, а не сами значения этих атрибутов. Поэтому, если мы передадим значение атрибута, обработки не будет. Мы можем обрабатывать такие случаи, но чаще всего нам легче получить значение атрибута класса enum немного иначе.

```python
processing_image(b'webp', ImageExtension['webp'.upper()])
```

```python
from enum import auto, Flag
class Weekday(Flag):
    MONDAY = auto()
    TUESDAY = auto()
    WEDNESDAY = auto()
    THURSDAY = auto()
    FRIDAY = auto()
    SATURDAY = auto()
    SUNDAY = auto()
    WEEKEND = SATURDAY | SUNDAY
```

Ну, тут стоит знать, что defaultdict быстрее и правильнее работает с точки зрения создания ключей, чем set-default. ## Копирование объектов

```python
import copy
```

```python
a = [1,2,3]
b = a
print(a==b)
b is a
```

```python
c = copy.copy(a) # a.copy()
```

```python
print(a == c)
print(c is a)
print(id(c), id(a))
c.append(4)
print(a,c)
```

```python
lst = [1,2,3,[1,2,3], [[4]]]
```

```python
lst1 = copy.copy(lst)
print(lst == lst1)
print(lst is lst1)
print(lst[3] is lst1[3])
```

```python
lst2 = copy.deepcopy(lst)
```

```python
print(lst == lst2)
print(lst is lst2)
print(lst[3] is lst2[3])
```

## Fuctools partial

Этот декоратор используется для создания новых функций из существующих с уже задаными аргументами

```python
def triangle_aray(a,b,c, add=0, mul=1):
    print(a,b,c, add, mul)
    return (a + b + c) * mul + add

res = triangle_aray(1,2,3)
```

```python
res
```

```python
from functools import partial
```

```python
rect_per = partial(triangle_aray, 0, 1)
rect_per(6, mul=2)
rect_per(1,2,5)
rect_per(a=1,b=2,c=3)
```

```python
rect_per = partial(triangle_aray, b=0, c=1)
rect_per(6)
```

## cache и lru_cache

```python
from functools import cache, lru_cache

@cache
def factorial(n):
    print('Call')
    return n * factorial(n-1) if n else 1
```

```python
factorial(10)
```

```python
factorial(5)
```

```python
factorial.cache_info()
```

```python
factorial.cache_parameters()
```

```python
factorial(12)
```

```python
@lru_cache(maxsize=5,typed=True)
def factorial_new(n: int):
    print('Call')
    return n * factorial_new(n-1) if n else 1
```

```python
help(factorial_new)
```

```python
factorial_new(1)
```

```python
factorial_new.cache_clear()
```

```python
factorial_new(10)
```

```python
factorial_new(5)
```
