## Пару слов о выражениях

Любая программа - это по сути своей текст, который придерживается определённого формата


Текст python программ состоит из `expressions`. Внутри expression используются `operators` (операторы) и `operands` (операнды).


Python для каждого expression’а будет совершать набор действий. Совершение этих действий называется `«evaluation»`.

После evaluation остаётся `value` (результат). По-русски результат. Но давайте опять оставим английский термин.

В итоге:
- **Expressions**: минимальные детали, которые Python может распознать в тексте вашей программы.
- **Evaluation**: набор действий, которые совершает Python для каждого expression’а при выполнении программы.
- **Value**: штука, которая остается каждый раз после выполнения evaluation для expression (за исключением пары специальных случаев).

**Важно сказать про порядок выполнения операторов (приоритеты)**

* Группа 1, выполняются первыми

    – ** возведение в степень

* Группа 2

    – + признак положительного числа для одного аргумента, например +1
    – - признак отрицательного числа для одного аргумента, например -2

* Группа 3

    – * умножение
    – / деление
    – // целочисленное деление (не путать с комментарием в C++ и Java)
    – % остаток от деления (потом расскажу, что это такое)

* Группа 4, выполняются последними

    – + сложение двух аргументов
    – - вычитание двух аргументов


Операторы () не имеют приоритетов. После эвалуэйта этого оператора остаётся value, как после эвалуэйта всей цепочки экспрешенов внутри него. И в python, как и в матиматике, просто принят принцип - что сначала эвалуэйтится оператор скобок и оставляет после себя value, к которому дальше применяются операторы

Что ещё умеем делать со списками?

* pop(num) - удали элемент и верни его (num - с какой позиции забрать, если не указываем, то будет последний)

* remove(elem) - удаляем по элементу. Берет первое вхождение и удаляет его

* clear() - очистить весь список

* len(l) - вывести размер списка (сколько элементов содержит)

* count(elem) - посчитай, сколько раз элемент входит в список

* index(elem) - выведи индекс первого вхождения элемента (если элемента нет, то выведет ошибку)

* reverse() - переверни список

* sort(reverse=False) - отсортируй список (по возрастанию, если поставить reverse=True, то будет по убыванию). Аналогично есть функция sorted(), которая делает то же самое, но только не меняет список

* max/min(a) - найти максимум/минимум в списке

#### Множества (Set)

**Множество** — cоставной тип данных, представляющий собой <u>неупорядоченный</u> набор уникальных объектов (элементов множества) под одним именем. Множества изменяемы и чаще всего используются для удаления дубликатов и всевозможных проверок на вхождение.

````python
my_set = set()

my_set.add("hello")
my_set.add("hello")
my_set.add("hello")
my_set.add("my friend!")

print(my_set)
````

Для удаления есть три метода: \
`.discard()`  - удалить заданный элемент, если он есть в множестве, и ничего не делать, если его нет; \
`.remove()` - удалить заданный элемент, если он есть, и породить ошибку `KeyError`, если нет; \
`.pop()` - удалить некоторый элемент из множества и возвратить его как результат.

Очистить множество от всех элементов можно с помощью метода <b>clear</b>:

````python
my_set.clear()
````

Конечно, здесь есть и другие элементы из теории множеств. Например, операции над множествами:

### Match case

Вообще `if` ... `elif` ... `else` можно использовать реже, ведь есть `match` ... `case`! Привет python 3.10)

````python
exchanger: str = input("Каким файлообменником ты пользуешься? ")

match(exchanger):
    case "Torrent":
        print("Тогда пиров и сидов тебе побольше")
    case "Yandex" | "Яндекс" | "disk.yandex.ru":
        print("Хе-хе, мемно, что у них 'disk' в доменном имени, да?")
    case _:
        print("Ой, я такого не знаю. Могу порекомендовать Skype. Он так всё сохранил, что даже я открыть теперь не могу.")
````

Основные шаблоны, которые можно использовать в case:

* Literal pattern ("Torrent") — точное совпадение. Используется, когда нужно проверить конкретное значение.
* Wildcard pattern (_) — по умолчанию. Срабатывает, если ни один другой шаблон не подошёл.
* Sequence pattern — структура списка или кортежа. Используется для сопоставления с фиксированной структурой (порядок и количество элементов).
* Mapping pattern — структура словаря. Применяется, когда нужно проверить наличие нужных ключей и их значений.
* Capture pattern — захват значений. Позволяет сохранить значение в переменную.
* Class pattern — объекты и классы. Используется для сопоставления значений внутри объектов по их полям.

````python
def process_coordinates(point):
    match point:
        case (0, 0):
            print("Точка в начале координат")
        case (x, 0):
            print(f"Точка на оси X: {x}")
        case (0, y):
            print(f"Точка на оси Y: {y}")
        case (x, y):
            print(f"Точка в ({x}, {y})")
        case _:
            print("Не координаты")

process_coordinates((0, 1))
````

````python
def process_items(items):
    match items:
        case []:
            print("Пустой список")
        case [first]:
            print(f"Один элемент: {first}")
        case [first, second]:
            print(f"Два элемента: {first} и {second}")
        case [first, *rest] if len(items) < 5:
            # Обратите внимание на доп.условие.
            print(f"Короткий список. Первый элемент: {first}, остальные: {rest}")
        case [first, *rest]:
            print(f"Длинный список. Первый элемент: {first}, остальные: {rest}")

process_items([1, 2, 3, 4])
````

Замечание про связный список

Трушный LinkedList в Python придётся делать самому. В Python `deque` работает по схожему принципу, вставка и удаление на концах работает за константное время. Но вот удалять произвольные элементы (на на концах) нельзя.

````python
from collections import deque

# Создание deque
d = deque([1, 2, 3])

# Добавление в конец (как в обычный список)
d.append(4)  # deque([1, 2, 3, 4])

# Добавление в начало
d.appendleft(0)  # deque([0, 1, 2, 3, 4])

# Удаление с конца
d.pop()  # 4, deque([0, 1, 2, 3])

# Удаление с начала
d.popleft()  # 0, deque([1, 2, 3])
````
