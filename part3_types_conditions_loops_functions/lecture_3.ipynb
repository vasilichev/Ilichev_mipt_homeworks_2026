### Параметры, аргументы и возвращаемые значения

Как можно комбинировать callable? если каждая из них при вызове использует уникальную область видимости, и мы не можем сделать evaluation для идентификаторов из «соседней» callable?

````python
# функция с аргументами

# (4) Python начинает использовать новый scope
# (5) Python указывает идентификатор "operand" на value 10
def square(operand):
    # (6) Python использует идентификатор "operand" из scope
    operand * operand
# (1) Python выполняет evaluation для литерала 10, остается value
# (2) Python запоминает, что первый аргумент - это value 10
# (3) Python вызывает callable square
square(10)

square(10) # 10 - это аргумент функции
````

````python
# функции могут иметь больше одного параметра
def area(width, height):
    width * height
````

Для передачи value из области видимости callable в Python предусмотрен «return statement». Так же, как уже знакомый вам «function definition statement», return — это statement. То есть «избранный» expression, который должен располагаться на отдельной строке и не может быть частью других expression и statement. Зато другие expression’ы (но не statement’ы!) могут быть его частью.

````python
# (4) Python начинает использовать новый scope
# (5) Python указывает идентификаторы width и height на value 12 и 2
def area(width, height):
    print(id(width), id(height))
    # (6) Python выполняет evaluation для "width * height", остается value
    # (7) Python запоминает, что полученное value 24 - это результат callable
    return width * height
# (1) Python выполняет evaluation для литералов, остаются 2 value
# (2) Python запоминает, что эти value - первый и второй аргументы
# (3) Python вызывает callable area
width = 15
result = area(width, 2)
# (8) Python указывает идентификатор result на value 24
result
````

### Области видимости

Когда мы используем имя в программе, Python создаёт, изменяет или ищет это имя в том, что называется пространством имён — месте, где живут имена. И когда мы говорим об именах в отношении к коду, пространства имён соответствуют областям видимости: местоположение присваивания имени в вашем исходном коде определяет область видимости видимости имени в вашей программе.

Место, где вы присваиваете имя в исходном коде, определяет пространство имён, в котором оно будет жить, а значит и область видимости этого имени.

Кроме упаковки кода для повторного использования, функции добавляют дополнительный уровень пространства имён в ваши программы, чтобы минимизировать вероятность конфликтов между переменными с одинаковыми именами — по умолчанию все имена, присвоенные внутри функции, связаны только с пространством имён этой функции.

В python существуют специальные правила по разделению областей видимости. 



````python
x = 24 # global

def outer():
    x = 42 # enclosing

    def inner():
        x = 12 # function

    inner()
    print("New x = ", x)

outer()
````

### Сопоставление аргументов

- **Позиционные**: сопоставляются слева направо\
Обычный случай, который мы в основном использовали до сих пор — сопоставить значения аргументов (переданные при вызове) с именами аргументов (указанными в определении функции) по позиции, слева направо.

- **Именованные**: сопоставляются по имени аргумента\
В качестве альтернативы вызывающие могут явно указать, какой аргумент функции должен получить значение, используя синтаксис имя=значение с именем аргумента из определения.

- **Значения по умолчанию**: задают значения для необязательных аргументов, которые не передаются\
Функции могут сами задавать значения по умолчанию для аргументов, которые будут использоваться, если при вызове передано слишком мало значений, также с использованием синтаксиса имя=значение.

- **Звёздчатые сборщики (args)**: собирают произвольное количество позиционных или именованных аргументов\
Определения функций могут использовать специальные аргументы с одним или двумя символами * перед именем для сбора произвольного числа аргументов после другого сопоставления. Эта функция иногда называется varargs (переменное число аргументов) по аналогии с аналогичным инструментом в языке С; в Python аргументы собираются в обычный объект.

- **Звёздчатые распаковщики(kwargs)**: передают произвольное количество позиционных или именованных аргументов\
Вызовы функций также могут использовать один или два символа * для распаковки коллекций аргументов в отдельные аргументы. Это обратная операция к * в определении функции: в определении это означает собирать произвольное число аргументов, а в вызове — распаковывать произвольное число аргументов и передавать их индивидуально как отдельные значения.

- **Аргументы только по ключу**: аргументы, которые должны быть переданы по имени\
Определения функций могут также использовать * в заголовках для указания аргументов, которые должны быть переданы только по имени как именованные аргументы, а не позиционно. Это часто используется для параметров конфигурации, дополняющих основные аргументы.

- **Позиционные только аргументы**: аргументы, которые должны быть переданы по позиции\
Начиная с Python 3.8, определения функций могут дополнительно использовать / в списке аргументов заголовка, чтобы указать, что все аргументы, предшествующие этому символу, должны быть переданы по позиции, а не по имени именованного аргумента.Если нужна помощь с дальнейшим переводом или пояснениями по синтаксису позиционных только аргументов в Python, обращайтесь.

````python
def func(name): ...  # Позиционный аргумент
def func(name=value): ...  # Именованный аргумент и значение по умолчанию
def func(id_, *name): ...  # Функция, Сопоставляет и собирает переданные, кроме позиционных и именованных, позиционные аргументы в кортеж
def func(**name): ...  # Функция, Сопоставляет и собирает оставшиеся позиционные аргументы в словарь
def func(*, name): ...  # Аргументы функции, которые должны быть переданы в вызовах только по ключу
def func(name): ...  # Аргументы функции, которые должны быть переданы в вызовах только позиционно
````

Когда эта функция вызывается, Python собирает все позиционные аргументы в новый кортеж и присваивает переменной args этот кортеж. Поскольку это обычный объект кортежа, по нему можно индексировать, проходить циклом for и так далее

С позиционными аргументами всё обстоит похожим образом: он собирает аргументы в новый словарь. А потом можно работать с kwargs как со словарём

````python
def f(**kwargs):
    print(kwargs)

f()

f(a=1, b=2)
````

Схема передачи по присваиванию Python — не совсем то же самое, что и ссылочные
параметры C++, но на практике она очень похожа на модель передачи аргументов
в С (и других языках).
- Неизменяемые аргументы фактически передаются “по значению”\
    Объекты, подобные целым числам и строкам, передаются по ссылке на объекты, а не путем копирования, но из-за того, что модифицировать на месте неизменяемые объекты невозможно, эффект во многом похож на создание копий.
- Изменяемые аргументы фактически передаются “по указателю”\
    Объекты вроде списков и словарей также передаются по ссылке на объекты, что аналогично способу передачи массивов как указателей в языке С — изменяемые объекты можно модифицировать на месте в функции почти как массивы в С

````python
# более сложный пример с разными типами аргументов

def test(name, *, test_1, test_2):...

def tag(name, *content_data, class_=None, id_=None, **attrs):  # хак с class_ нужен, т.к class - это зарезервированное имя в python
    if class_ is not None:
        attrs['class'] = class_
    if id_ is not None:
            attrs['id'] = id_

    attr_pairs = ''.join(f' {attr}="{value}"' for attr, value in attrs.items())
    if content_data:
        return '\n'.join(f'<{name}{attr_pairs}>{c}</{name}>' for c in content_data)
    return f'<{name}{attr_pairs} />'
````

````python
print(tag('br'))
print(tag('p', 'hello'))
print(tag('p', 'hello', 'world'))
print(tag('p', 'hello', id_=123))
print(tag('p', 'hello', 'world', class_='sidebar'))
print(tag(content='testing', name='img'))
my_tag = {'name': 'img', 'title': 'Sunset', 'src': 'sunset.jpg', 'class_': 'framed'}
print(tag(**my_tag))
print(tag('img', id_=678, 'o00'))
````

### Чистые функции

**Детерминированность функций** — это свойство, при котором функция всегда возвращает один и тот же результат для одних и тех же входных данных без побочных эффектов, связанных с состоянием программы. Детерминированная функция не зависит от контекста и внешних факторов.

Рассмотрим пример детерминированной функции:

```python
def multiply(a, b):
    return a * b
```
Нельзя однозначно сказать, что отсутствие детерминированности — это плохо. Для работы многих программ и сайтов нужна функция, которая будет возвращать случайное число или вычислять текущую дату. Но мы можем разделить код так, чтобы в нем было как можно больше детерминированных частей.

**Побочные эффекты или side effects** — это любые взаимодействия с внешней средой. К ним относятся файловые операции, такие как запись в файл, чтение файла, отправка или приём данных по сети и даже вывод в консоль.

Побочные эффекты составляют одну из самых больших сложностей при разработке. Они затрудняют логику кода и тестирование. Это приводит к большому числу ошибок. Например, только при работе с файлами количество возможных ошибок измеряется сотней. В этом случае может не хватить места на диске или будет попытка прочитать данные из несуществующего файла. Для предотвращения ошибок код обрастает большим числом проверок и защитных механизмов.

**Чистые функции или pure functions** — это функции, которые при вызове не влияют на состояние программы и не имеют побочных эффектов. Они возвращают значения только на основе входных аргументов и не изменяют их.

```python
def add_numbers(x, y):
    return x + y


# создаем новый список вместо изменения существующего
def add_user(users, new_user):
    return [*users, new_user]

```

### Элементы функционального программирования.

В функциональных языках очень часто есть ФВП map,filter и reduce (имена могут различаться).

В python тоже map и filter являются встроенными (built-in) функциями, но они уже немного потеряли актуальность, т.к у нас есть генераторные выражения и списковые включения
Давайте сравним

````python
def factorial(n):
    return 1 if n < 2 else n * factorial(n-1)
````

````python
list(map(factorial, range(6)))
````

````python
[factorial(n) for n in range(6)]
````

````python
list(map(factorial, filter(lambda n: n % 2, range(6))))
````

````python
[factorial(n) for n in range(6) if n%2]
````

При этом стоит помнить разницу, что map и filter возвращают генераторы (разберём далее), поэтому если писать прям замену, то стоит использовать генераторное выражение `(...)`, а не списковое включение

### Модули itertools, functools и operator

С концепцией разбиения на модули и пакеты мы познакомимся на последующих лекциях, но сейчас хочу показать разные функциональные проявления в языке

**Operator**

Для того, чтобы писать в функциональном стиле очень часто бывает удобно представлять стандартные операции в виде функций.

Самый банальный пример - вычисление факториала от числа.

````python
# без модуля operator

from functools import reduce

def factorial(n):
    return reduce(lambda a,b: a*b, range(1, n+1))
````

````python
# Чтобы не писать lambda и подобные операторы

from operator import mul

def factorial_func(n):
    return reduce(mul, range(1, n+1))
````

````python
factorial(5)
````

````python
factorial_func(5)
````

Мы с вами говорили о том, что можно получать доступ к элементам последовательности по индексу с помощью `[]`

В модуле operator есть функциональная реализация itemgetter. Давайте посмотрим пример

````python
from operator import itemgetter

cities = [
    ("New York", "NYC", 27000, (40.7128, -74.0060)),
    ("New York", "NYC", 7000, (40.7128, -74.0060)),
    ("London", "LDN", 18000, (51.5074, -0.1278)),
    ("Tokyo", "TYO", 23000, (35.6895, 139.6917)),
    ("Berlin", "BER", 9000, (52.5200, 13.4050)),
    ("Sydney", "SYD", 12000, (-33.8688, 151.2093)),
    ("Moscow", "MOW", 15000, (55.7558, 37.6173))
]

def sort_with_sugar():
    for city in sorted(cities, key=lambda c: (c[1], c[2])):
        print(city)

def sort_with_func():
    for city in sorted(cities, key=itemgetter(1,2)):
        print(city)
````

````python
sort_with_sugar()
````

````python
sort_with_func()
````

# Аннотации типов

Зачем нужно указывать типы?
* Помогают раньше находить ошибки в коде - IDE и type checkers нам в помощь
* Помогают разработчику понять, что передавать в функцию и что ожидать на выходе

Помним, что Python сам по себе не запретит запустить нам код с неверными типами. Ошибку нужно ловить через type checkers!

````python
from typing import (
    NamedTuple,
    List,
    Dict,
    Set,
    Tuple,
    Optional,
    Union,
    Literal,
    Callable,
    Iterable,
    Sequence,
    Mapping,
    Generator,
)
````

* Начиная с Python 3.10 List, Dict, Set, Tuple можно (и нужно) заменять на встроенные типы: list[int], dict[str, str], set[str], tuple[int, ...]
* `Optional[str]` <=> `str | None`
* `Union[str, int]` <=> `str | int`
* `Literal["gray", "color"]` <=> можно передать только строку "gray" или строку "color"
---
* `Iterable` - Всё, по чему можно итерироваться: словари, списки, кортежи...
* `Sequence` - Всё, где допустимо обращение по индексу: список, кортеж, строка...
* `Mapping` - Всё, где есть обращение по ключу: словарь, кастомные типы...

````python
from dataclasses import dataclass
from enum import auto, Enum, StrEnum


class GenderCompatible(Enum):
    MALE = 1
    FEMALE = 2
    UNISEX = auto()


class ShirtMaterial(StrEnum):
    COTTON = "Хлопок"
    VISCOSE = "Вискоза"
    SYNTHETICS = "Синтетика"


class ShirtColor(StrEnum):
    BLACK = "Чёрный"
    WHITE = "Белый"
    BLUE = "Синий"
    RED = "Красный"

class JeansColor(StrEnum):
    BLACK = "Чёрный"
    WHITE = "Белый"
    BLUE = "Синий"
    RED = "Красный"

class ShirtSize(StrEnum):
    S = "S"
    M = "M"
    L = "L"
    XL = "XL"
    XXL = "2XL"
    XXXL = "3XL"


class Price(NamedTuple):
    price: float
    currency: str

@dataclass
class Shirt:
    price: Price
    gender: GenderCompatible
    material: ShirtMaterial
    color: ShirtColor
    size: ShirtSize
    tags: tuple[str, ...]
    description: str | None = None


@dataclass
class Jeans:
    price: Price
    brand: Literal["Montana", "Not Montana"]
    color: JeansColor


def get_jeans_info(jeans: Jeans) -> str:
    return f"Джинсы от бренда {jeans.brand}\n"


def get_shirt_info(shirt: Shirt) -> str:
    return (
        f"Материал: {shirt.material.value}\n"
        f"Размер: {shirt.size.value}\n"
    )


def get_card_info(good: Shirt | Jeans) -> str:
    info = f"Цена: {good.price.price} {good.price.currency}\nЦвет: {good.color.value}\n"
    if isinstance(good, Jeans):
        info += get_jeans_info(good)
    elif isinstance(good, Shirt):
        info += get_shirt_info(good)
    return info


goods = [
    Shirt(
        price=Price(5000, "RU"),
        gender=GenderCompatible.MALE,
        material=ShirtMaterial.COTTON,
        color=ShirtColor.BLACK,
        size=ShirtSize.XXL,
        tags=("Деловой стиль", "Лимитированная коллекция"),
    ),
    Shirt(
        price=Price(10000, "RU"),
        gender=GenderCompatible.FEMALE,
        material=ShirtMaterial.SYNTHETICS,
        color=ShirtColor.WHITE,
        size=ShirtSize.XL,
        tags=("Кэжуал",),
        description="Замечательный фэбрик"
    ),
    Jeans(
        price=Price(4000, "RU"),
        brand="Montana",
        color=JeansColor.BLUE,
    ),
    Shirt(
        price=Price(1500, "RU"),
        gender=GenderCompatible.UNISEX,
        material=ShirtMaterial.SYNTHETICS,
        color=ShirtColor.RED,
        size=ShirtSize.XXL,
        tags=("Кэжуал",),
    ),
    Jeans(
        price=Price(8000, "RU"),
        brand="Not Montana",
        color=JeansColor.WHITE,
    ),
]
````

````python
for good in goods:
    print(get_card_info(good))
````

````python
import random


def get_top(
    goods: Sequence[Shirt | Jeans],
    sort_condition: Callable,
    limit_count: int | None = None,
    order: Literal["asc", "desc", "shuffle"] = "asc",
) -> Iterable[Shirt | Jeans]:
    match order:
        case "shuffle":
            sorted_goods = goods[:]
            random.shuffle(sorted_goods) # type: ignore
        case "desc" | "asc":
            sorted_goods = sorted(goods, key=sort_condition, reverse=order=="desc")
        case _:
            raise ValueError("Unknown flag!")
    if limit_count is not None and limit_count >=0:
        return sorted_goods[:limit_count]
    return sorted_goods
````

````python
get_top(goods, sort_condition=lambda x: x.price.price, order="desc", limit_count=-1)
````
