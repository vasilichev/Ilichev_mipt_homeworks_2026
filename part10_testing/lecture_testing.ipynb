# Лекция: Тестирование в Python

> _"Тестирование может показать присутствие ошибок, но не их отсутствие"_
>
> **Эдсгер Дейкстра**

Иными словами тестирование может помочь найти ошибки, но не гарантировать, что в коде их нет.

Если все тесты прошли, это не значит, что ошибок нет!

# Небольшой глоссарий

**Тестирование** - это проверка соответствия поведения проверяемой системы требованиям, выполняемая по результатам реальной работы этой системы в некотором <u>конечном</u> наборе специально созданных ситуаций.

**Качество ПО** - пригодность ПО для решения задач

Согласно ISO 25010:2011 к характеристикам качества ПО относят:


<details>
<summary>
<b>Функциональность</b>
</summary>

Требование должно содержать: описание входных данных и условий, выполняемых преобразований и действий, результатов и их характеристик

Пример: "Калькулятор должен вычислять sin на диапазоне от 10^-300 до 10^300 с точностью 10^-20"
</details>


<details>
<summary>
<b>Производительность</b>
</summary>

Расход всех вычислительных ресурсов при решении задач: CPU, RAM, пропускная способность каналов связи...

Пример: При работе 10 пользователей средняя загрузка процессора основного сервера не должна превышать 80%. Нагрузка на каналы связи не должна быть выше 20 МБ/с. При построении отчёта затраты ОЗУ не должны превышать 2GB.
</details>

<details>
<summary>
<b>Надёжность</b>
</summary>

Работоспособность системы в некорректном окружении, а также зрелость самой системы.
<ul>
  <li>Работа с некорректными входными данными</li>
  <li>Продолжение работы/сохранение данных при сбоях электропитания</li>
  <li>Продолжение работы/сохранение данных при сбоях каналов связи (слышали про offlne-first приложения?)</li>
  <li>Среднее ожидаемое количество сбоев в заданном интервале времени</li>
  <li>Среднее время и трудоёмкость полного восстановления работоспособности при серьёзном сбое</li> 
</ul>
</details>

<details>
<summary>
<b>Переносимость</b>
</summary>

Работоспособность в другом окружении (другие платформы, другие форматы данных) + возможность замены другого ПО этим.

Пример:
Работает ли программа, созданная для Windows 10, в Windows 11?

Можно ли заместить Microsoft Office при помощи нашего форка Libre Office?
</details>

<details>
<summary>
<b>Совместимость</b>
</summary>

Способность к взаимодействию/сосуществованию с другими системами того же уровня.

Пример: может ли СУБД Oracle работать с SAP R/3 на одном сервере?
</details>

<details>
<summary>
<b>Защищённость</b>
</summary>

<ul>
  <li><b>Конфиденциальность</b> - невозможность прочесть данные, для чтения которых нет прав</li> 
  <li><b>Целостность</b> - невозможность модицировать данные и влиять на выполнение операций, если нет прав</li>
  <li><b>Аутентичность</b> - способность подтвердить, что данные или компоненты системы не подвергались модификациям, за исключением доверенных</li>
  <li><b>Авторизуемость</b> - способность подтверждаемым образом привязывать к каждому действию пользователя, от имени которого оно выполнено.</li>
</ul>
</details>

<details>
<summary>
<b>Удобство использования</b>
</summary>

Производительность пользователей при решении задач с помощью разработанной системы + производительность при обучении.

Такое плохо формализуется - часто используются экспертные оценки и специальные метрики (замеры времени для различных сценариев и профилей, время удержания)
</details>

<details>
<summary>
<b>Удобство сопровождения</b>
</summary>

Производительность разработчиков при решении задач сопровождения и развития системы.

<ul>
  <li>Удобство тестирования и покрытие тестами</li> 
  <li>Возможность изменения модуля без влияния на остальные</li>
  <li>Вероятность создания новой ошибки при внесении изменений</li>
</ul>
</details>


**Методы контроля качества ПО**:
- **Верификация** (соответствует ПО требованиям или нет?)
- **Валидация** (соответствует конечным целям или нет?)

## Цели тестирования

- Проверка правильности реализации
- Проверка обработки внештатных ситуаций и граничных условий
- Минимизация последствий (ошибок)
- Подготовка ко внесению изменений (например, рефакторинг)

Давайте разбираться с уровнями тестирования:

**Unit-тесты**: берут маленький кусочек продукта (метод, класс) и проверяют его изолированно. Это быстро, надёжно, ошибки легко локализуюся, такие тесты менее сложны в реализации и их можно делать много. Не нужно разворачивать всю систему.

**Интеграционное тестирование**: проверятся интеграция модулей, их взаимодействие друг с другом, интеграция подсистем. Например, тестируем добавление записи товара + проверяем факт создания в базе. Такие тесты "дороже", сложнее, но позволяют проверить совместную работу компонентов.

**Системное тестирование**: иногда выделяют, иногда нет. Проверям, что система соответствует функциональным и нефункциональным требованиям, но можем использовать для этого отдельную песочничу (используем stage-стенды других сервисов, что-то мокируем).

**End-2-End**: тесты симулируют поведение пользователя в реальной системе. Например, пользователь заходит на сайт, входит в аккаунт, просматривает ленту новостей и ставит лайк. Ещё дороже, нужна полностью развёрнутая система, баги сложнее локализовать. Никаких заглушек.

Более подробно про изменения:

**Regression** - проверка всего существующего функционала (проверяем, что функционал не сломан)

**Sanity** - проверка конкретного функционала

**Build verification** - проверка критически важного функционала на новой версии

# Немного о тест-дизайне

Так как исчерпывающее тестирование в реальной жизни почти невозможно, нужно грамотно составлять тесты, чтобы закрыть бОльшую часть потенциальных дыр. Стараемся избегать дублей, избыточности и однотипных проверок.

## Классы эквивалентности

Разделяем функционал или данные на определенные наборы (классы эквивалентности), с которыми тестируемое приложение должно работать одинаково. То есть предполагается, что система обрабатывает все значения из одного и того же класса одинаковым образом. Следовательно, для проверки поведения системы достаточно выбрать по одному элементу из каждого класса.

<br>

Пусть есть поле для ввода возраста пользователя на форме регистрации.

Требование: **"Возраст должен быть в диапазоне от 18 до 59 лет включительно".**

На основе этого требования можно выделить следующие классы эквивалентности:

- КЭ1 (невалидный): возраст <18 (например, 17, 0, -5)
- КЭ2 (валидный): 18≤ Возраст ≤59 (например, 25, 18, 59)
- КЭ3 (невалидный): Возраст >59 (например, 60, 100)

Для тестирования достаточно выбрать по одному значению из каждого класса, например: 17, 35 и 60. Это позволяет тремя тест-кейсами покрыть весь спектр возможного поведения системы, связанного с этим полем, вместо тестирования сотен возможных значений.

## Граничные значения

Это те места, в которых один класс эквивалентности переходит в другой.

Достаточно проверить следующие значения:
- На 1 ниже границы
- На 1 выше границы
- Минимальное
- Максимальное

## Pairwise testing

Попарное тестирование - техника тестирования методом черного ящика, при которой тестовые сценарии разрабатываются таким образом, чтобы выполнить тестирование для всех комбинаций параметр-значение.

Интересный кликабельный пример [здесь](https://www.tester-today.com/pairwise-testing). Давайте представим, что есть форма регистрации с тремя выпадающими списками. В первом три опции, во втором две и в третьем тоже две. Всего вариантов 3\*2\*2 = 12. На самом деле это количество избыточно, достаточно просто покрыть попарные сочетания:

Однако парное тестирование не является универсальным решением для всех случаев тестирования. def calc_discount(age: int, is_active: bool) -> float:

```python
    if not is_active:
```

## Unittest (встроен в Python)

Основная сущность - класс `TestCase`, который будет группировать наши тесты.

class TestCase:

- def setUp(self): запускается перед каждым тестом

- def tearDown(self): очистка после каждого метода

- def test_<some_name>(self): сам тест

````python
import unittest

class TestMultiplication(unittest.TestCase):
    def test_multiply(self):
        self.assertEqual(1*2, 2)
````

# А как вообще устроен тест?

Структура теста: AAA

- **Arrange** – подготовка данных для теста
- **Act** – что проверяем в тесте (чем короче, тем лучше)
- **Assert** – проверка результатов полученных от действий


```python
def test_block(mock_urlopen: MagicMock) -> None:
    """Проверка блокировки автомата"""

    mock_urlopen.side_effect = TimeoutError()
```

# Test Driven Development (TDD)

Ссылка на оригинальный обзор - [тут](https://www.thedigitalcatonline.com/blog/2020/09/10/tdd-in-python-with-pytest-part-1/).

The TDD mantra is *"Test first, code later"*

Требования -> Тесты -> Код

TDD медленнее, чем разработка без тестов, но время, инвестированное в написание тестов обычно сравнимо меньше, чем количество времени, которое будет тратиться на поиск и исправления багов.

Код не сразу должен быть идеальным, он лишь должен проходить тесты, потом должен последовать рефакторинг. В теории, рефакторинг не должен добавлять новое поведение, должен являть собой идемпотентную трансформацию. 

## 6 заветных правил TDD

**Правило №1** - Сначала тесты, потом код.

**Правило №2** - Добавляем только минимально необходимый код, чтобы пройти тесты.

**Правило №3** - В момент времени не должно быть более одного упавшего теста.

**Правило №4** - Напиши код, который проходит тесты. Потом делай рефакторинг.
Иными словами, код может быть сколь угодно ужасен, но если он проходит тесты, мы считаем его корректным. Для проверки качества можно использовать другие виды тестирования (например, тестирование производительности)

**Правило №5** - Тест должен падать при первом запуске. Если этого не случилось, надо убедиться, что был смысл его добавлять.

**Правило №6** - Никогда не рефактори код без тестов.

Пусть наша цель - написать класс `SimpleCalculator`, который выполняет вычисления: сложение, вычитание, умножение и деление.
- Сложение и умножение должны принимать несколько аргументов.
- Деление должно возвращать значение с плавающей точкой, а деление на ноль должно возвращать строку "inf".
- Умножение на ноль должно вызывать исключение ValueError.
- Класс также предоставляет функцию для вычисления среднего значения для итерируемого объекта, такого как список. Эта функция получает два необязательных верхнего и нижнего пороговых значения и должна удалять из вычисления значения, которые выходят за эти границы.

#### Как pytest находит тесты:

1. Рекурсивно находит все python-файлы в текущей директории
2. Оставляет только файлы вида `test_*.py` и `*_test.py`
3. В этих файлах
- Находит все функции с префиксом `test`
- Находит все методы с префиксом `test` внутри классов с префиксом `Test`. У классов не должно быть метода `__init__`
  
Поведение можно модифицировать. [Подробнее в документации](https://docs.pytest.org/en/stable/goodpractices.html#test-discovery)

````python
%%ipytest -q
import pytest


@pytest.mark.parametrize(
    "first, second, expected",
    [
        (-1, 2, 1),
        (1, -2, 0),
        (3, 4, 7),
        (-5, -6, -11),
    ],
    ids=["first negative", "second negative", "both positive", "both negative"],
)
def test_sample_parametrization(first, second, expected):
    assert first + second == expected
````

Давайте рассмотрим что из себя оно представляет:

Конкректно у нас фигурирует тут: `pytest.mark.parametrize`
Декоратор, который позволяет параметризировать (использовать несколько наборов данных) на тестовую функцию.

Каждая комбинация значений будет автоматически передаваться в тестовую функцию и pytest выполнит для каждого набора тестовых данных.



```python

@pytest.mark.parametrize("<аргумент1>, <аргумент2>, <аргумент3>, ...", [<набор_Значений_для_аргумента_1>, <набор_значений_для_аргумента_2>, <набор_значений_для_аргумента_3>, ...]
ИЛИ [
    (<набор_Значений_для_аргумента_1>, <набор_значений_для_аргумента_2>, <набор_значений_для_аргумента_3>, ...),
    (<набор_Значений_для_аргумента_1>, <набор_значений_для_аргумента_2>, <набор_значений_для_аргумента_3>, ...),
    (<набор_Значений_для_аргумента_1>, <набор_значений_для_аргумента_2>, <набор_значений_для_аргумента_3>, ...)
])
def test_<название теста>(<аргумент1>, <аргумент2>, <аргумент3>, ...):

    <подготовка или демонстрация тестового сценария>

    assert <проверяемое выражение>

```


А вот [так](https://docs.pytest.org/en/7.1.x/example/parametrize.html#parametrizing-conditional-raising) можно параметризовать работу с исключениями:

````python
from contextlib import nullcontext as does_not_raise

import pytest


@pytest.mark.parametrize(
    "example_input,expectation",
    [
        (3, does_not_raise()),
        (2, does_not_raise()),
        (1, does_not_raise()),
        (0, pytest.raises(ZeroDivisionError)),
    ],
)
def test_division(example_input, expectation):
    """Test how much I know division."""
    with pytest.raises(ValueError):
        call_my_func()
    with expectation:
        assert (6 / example_input) is not None
````

# Заглушки

Когда мы пишем unit-тесты, нам необходимо изолированно проверить некоторый участок кода.

По-хорошему, в Unit-тестах не должно быть походов в сеть или в базу - так успех теста начинает зависеть от внешних факторов + растёт стоимость теста.

Чтобы решить эту проблему, в тестировании используются test doubles (заглушки). Существует несколько разновидностей:

* **Dummy** - пустой объект, который нужен только чтобы передать зависимость (сам по себе в тесте не используется).

* **Стабы (Stub)** - возвращает заранее заданные (захардкоженные) данные.

* **Шпион (Spy)** - записывает, как с ним взаимодействовали (сколько вызовов, с какими аргументами). Не обязан что-то имитировать, может просто проксировать запросы.

* **Моки (Mock)** - объект с заранее заданными ожиданиями по вызовам (имена, количество); тест падает, если контракт взаимодействия нарушен.

* **Фейки (Fake)** - рабочая, но упрощенная реализация зависимости (например, in-memory репозиторий вместо реальной БД).


Источник: xUnit Test Patterns - Refactoring Test Code by Gerard Mesazaros, 2007

# Работа с моками

Пусть у нас есть очень простой класс, который позволяет запрашивать данные о погоде:

````python
import os

import httpx
from dotenv import load_dotenv

load_dotenv()


class WeatherService:
    def __init__(self, api_key: str) -> None:
        self.api_key = api_key

    def get_temperature(self, city: str) -> float:
        response = httpx.get(
            "https://api.weatherapi.com/v1/current.json",
            params={"key": self.api_key, "q": city},
        )
        response.raise_for_status()
        data = response.json()
        return data["current"]["temp_c"]


def main() -> None:
    api_key = os.getenv("WEATHER_API_KEY", "")
    weather_service = WeatherService(api_key)
    temperature = weather_service.get_temperature("Amsterdam")
    print(f"The current temperature in Amsterdam is {temperature}°C.")


if __name__ == "__main__":
    main()
````

Вызовы к внешним API обычно тарифицируются, а также содержат ограничения на число запросов.

Хороший unit-тест должен мокировать такие запросы.

## Вариант 1: Monkeypatch

Monkeypatch встроен в pytest. Указываем его в аргументах теста - pytest сам передаст его при запуске (это фикстура).

Теперь пишем *monkeypatch.setattr("httpx.get", fake_get)* - теперь вызод функции `httpx.get` будет заменяться вызовом `fake_get`

````python
from weather import WeatherService
import pytest
from typing import Any

def test_get_temperature_with_monkeypatch(monkeypatch: pytest.MonkeyPatch) -> None:
    def fake_get(url: str, params: dict[str, Any]) -> Any:
        class FakeResponse:
            def raise_for_status(self) -> None: pass
            def json(self) -> dict[str, Any]:
                return {"current": {"temp_c": 19}}
        return FakeResponse()

    monkeypatch.setattr("httpx.get", fake_get)
    service = WeatherService(api_key="fake-key")
    temp = service.get_temperature("Amsterdam")
    assert temp == 19
````

Что здесь не очень хорошо?

Появился большой блок кода с патчем и отдельный класс для ответа.

## Вариант 2: MagicMock

MagicMock - это сущность пакета unittest. Документация [здесь](https://docs.python.org/3/library/unittest.mock.html).

Это продвинутая заглушка-шпион - помимо симуляции поведения она позволяет проверять количество вызовов и прочую информацию.

````python
from typing import Any
from unittest.mock import MagicMock, patch

import pytest

from weather import WeatherService


def test_get_temperature_with_mocking() -> None:
    mock_response = MagicMock()
    mock_response.raise_for_status.return_value = None

    # А можно добавить в мок выброс исключения:
    # mock_response.raise_for_status.side_effect = ConnectionError("API unavailable")

    mock_response.json.return_value = {"current": {"temp_c": 25}}

    with patch("httpx.get", return_value=mock_response) as mock_get:
        service = WeatherService(api_key="fake-key")
        temp = service.get_temperature("London")

        assert temp == 25
        mock_get.assert_called_once()
````

Более подробно про side_effect можно почитать [здесь](https://www.py4u.org/blog/python-side-effect-mocking-behavior-of-a-method/)

## Но как быть с дублированием моков в каждом тесте?

Нужно использовать фикстуры! Это большое преимущество pytest - есть много встроенных фикстур + можно создавать свои. Фреймворк автоматически вызовет их перед началом теста/тестов. [Список лежит здесь](https://docs.pytest.org/en/stable/fixture.html). Наиболее интересные:
* monkeypatch - временно можифицирует методы классов, модулей и т.д.
* testdir - создает верменную директорию для каждого теста, которую потом чистит

У фикстур может быть разный scope - например создваться на каждый тест, модуль или тред, запускающий тесты. [Подробнее в документации](https://docs.pytest.org/en/stable/fixture.html).

Scopes (скоупы) у фикстур бывают:
* function (по умолчанию)
* class -- внутри класса
* module -- создаваемый внутри модуля

````python
import pytest
from weather import WeatherService
from typing import Any
from unittest.mock import MagicMock

# @pytest.fixture(scope='function') - значение по умолчанию
# @pytest.fixture(scope='module')
@pytest.fixture
def weather_service(monkeypatch: pytest.MonkeyPatch) -> WeatherService:
    def fake_get(url: str, params: dict[str, Any]) -> Any:
        mock_response = MagicMock()
        mock_response.raise_for_status.return_value = None
        mock_response.json.return_value = {"current": {"temp_c": 25}}
        return mock_response

    monkeypatch.setattr("httpx.get", fake_get)
    return WeatherService(api_key="fake-key")


def test_fixture_usage(weather_service: WeatherService) -> None:
    assert weather_service.get_temperature("Paris") == 25
````

Уже классно, но пока заглушка реализует только один сценарий. Давайте её параметризуем:

````python
@pytest.fixture(scope='function')
def mock_weather_service(request, monkeypatch: pytest.MonkeyPatch) -> WeatherService:
    mock_response, should_raise = request.param

    def fake_get(url: str, params: dict[str, Any]) -> Any:
        mock_response = MagicMock()
        if should_raise:
            mock_response.raise_for_status.return_value = Exception("Service connection failed")
        else:
            mock_response.raise_for_status.return_value = None
            mock_response.return_value = mock_response
        return mock_response

    monkeypatch.setattr("httpx.get", fake_get)
    return WeatherService(api_key="fake-key")
````

````python
%%ipytest -q


@pytest.mark.parametrize(
    "mock_service, expected_message",
    [
        (({"current": {"temp_c": 25}}, False), "25C"),
        (({}, False), ""),
        ((None, True), "Service connection failed"),
    ],
    indirect=["mock_service"],
)
def test_weather_service(mock_service, expected_message):
    service = UserService(mock_db)
    result = service.get_user_status(user_id)
    assert result == expected_message
    assert mock_db.get_user.call_count == 1
````

Что мы тут сделали?
* создали фикстуру в которой происходит подготовка всего необходимого для возвращения значений
* немного поправили входные параметры в нашей параметризации: то есть мы прокинули туда Mock объект и прописали логику при которой будет вызываться новое поведение в `request.params`

Такой механизм называется "непрямая параметризация" - `indirect parametrization`

В `pytest` объект request автоматически передается в фикстуры, когда они вызываются.
Например, с его помощью можно получить параметры, переданные в фикстуру, или информацию о тестовой функции, которая используетcя оной.
Когда используется параметризация тестов с флагом `indirect`, параметры передаются через фикстуру. Внутри фикстуры они становятся доступными через `request.param`

Для БД фикстура может выглядеть примерно так:

````python
%%ipytest -q

class DBConnection:
    pass

class TestDB:
    def init_db(self):
        print("init db")

    def get_connection(self):
        return DBConnection()

    def shutdown(self):
        print("close db")

@pytest.fixture(scope="module")
def db_connection():
    db = TestDB()
    db.init_db()
    try:
        yield db.get_connection()
    finally:
        db.shutdown()

def test_db_1(db_connection):
    assert db_connection

def test_db_2(db_connection):
    assert db_connection
````

Заключение:

* Мок может меняться и быть в тесте, а в свою очередь фикстура имеет свою структуру и не может меняться
* Моки легко параметризировать и писать различные тест кейсы на каждый случай
* Легко "вшивать фикстуры в тесты и прокидывать тестовые аргументы в параметры фикстуры тем самым задавая им свое поведение
